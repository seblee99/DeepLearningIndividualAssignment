**1.1**. What is a metric? What is a loss? What is the difference between both, and how are they used in the training process?

A metric, also known as a performance metric, is an encompassing term used to describe a measurement that assess how well a machine learning model is performing. A common example of a performance metric is accuracy.

Loss, which is also known as cost or objective function, describe an error in a model's prediction. Since it measures the performance of a model, it is considered a metric. Usually, the loss is defined as the difference between the predicted value and the actual value of the training data.

The difference between a metric and a loss is that a metric is an emcompassing term that provides the overall performance of the model and includes loss, and a loss is a detailed numerical measurement of the performance of a model.

Both are used in the training process. Metric is used to assess how well the model performed during or after the training. Loss is used during the optimization process, where the loss function is minimized through optimization, reducing errors and ultimately improving the overall metric.

1.3. Explain the bias-variance tradeoff problem, and how it is linked with the concept of overfitting and underfitting.

Bias is the difference between the expected and the true value, while variance is defined as the deviation of the model predictions from the expected values. Bias-variance tradeoff is a challenge in machine learning that requires finding the right balance between the bias and variance. This is because a model with high bias and low variance leads to a simple model, resulting in underfitting. On the other hand, a model with low bias and high variance leads to a complex model, resulting in overfitting. Therefore, finding the right balance between these two factors are crucial in preventing either underfitting or overfitting.

1.5. Briefly explain the CNN architecture, and why it is more adapted to images than standard MLP.

A Convolutional Neural Network consists of three primary groups of layers: convolutional layers, activation layers, and pooling layers. Convolutional layers scan input images for patterns and features, allowing the network to detect complex features. Activation layers, as described before, are composed of activation functions that introduce a non-linear transformation to the outputs of convolutional layers. Lastly, pooling layers reduce the dimensions of the representation and maintain translational invariance.

The CNN is more adapted to images than standard MLP due to sparse interaction, parameter sharing, and equivariant representation. First, traditional neural networks, such as standard MLP, require every output to interact with every input unit. In the case of image processing, only small, meaningful features are desired, reducing memory requirements and improving statistical efficiency. Second, parameters in CNN are shared across the entire image, whereas MLP requires separate parameters for each function, making CNN extremely computationally efficient. Finally, CNN becomes less sensitive to the location of features, which is crucial for identifying patterns in different locations in images.

```python
from sklearn.model_selection import train_test_split

def get_split_indices(dataset_length, train_ratio=0.7):
    """
    Function which splits the data into training and validation sets.
    arguments:
        dataset_length [int]: number of elements in the dataset
        train_ratio [float]: ratio of the dataset in the training set
    returns:
        train_indices [list]: list of indices in the training set (of size dataset_length*train_ratio)
        val_indices [list]: list of indices in the validation set (of size dataset_length*(1-train_ratio))
    """
    # ****START CODE****
    indices = np.arange(dataset_length)

    train_indices, val_indices = train_test_split(indices, train_size=train_ratio, shuffle=True)

    return train_indices, val_indices
    # ****END CODE****
```

2.3. Write the dataset classes. Feel free to add any type of data augmentation that you like. Please note that pytorch has an implemented PCAM dataset class, but we ask you to code these using from scratch. **[2 pt]**

```python
from torchvision import transforms
import matplotlib.pyplot as plt
import torch

class PCAMDataset(Dataset):
    def __init__(self, data, labels, train):
        """
        Dataset class for the PCAM dataset.
        arguments:
            data [numpy.array]: all RGB 96-96 images
            labels [numpy.array]: corresponding labels
            train [bool]: whether the dataset is training or validation
        """
        super(PCAMDataset, self).__init__()
        self.data = data
        self.labels = labels
        self.train = train

        if self.train:
            # ****START CODE****
            self.augmentation = transforms.Compose([
                transforms.ToTensor(),
                transforms.RandomHorizontalFlip(),
                transforms.RandomVerticalFlip(),
                transforms.RandomRotation(90),
            ])

            # ****END CODE****

    def __len__(self):
        # ****START CODE****
        '''
        Returns the length of the dataset
        '''
        return len(self.data)
        # ****END CODE****

    def __getitem__(self, idx):
        # ****START CODE****
        '''
        Returns image and target of the dataset
        '''

        image = (self.data[idx])
        label = int(self.labels[idx])

        if self.train:
            image = self.augmentation(image)
            # image = transforms.ToTensor()(image)
        else:
            image = transforms.ToTensor()(image)

        return image, label

        # ****END CODE****

    def display_sample(self, label=0):
        '''
        Returns a random image of declared label value from the dataset
        Input:
            label: 0 indicates negative and 1 indicates positive
        '''

        indices = np.where(self.labels == label)[0]
        index = np.random.choice(indices, replace=False)

        image, target = self[index]
        plt.imshow(np.asarray(image))
        plt.show()
        print(f"Label: {target}")
        print(f"Index: {index}")

    def display_random_sample(self):
        '''
        Returns a random image of undeclared label value from the dataset
        '''
        index = np.random.choice(len(self.labels), replace=False)
        print(f"Index: {index}")

        image, target = self[index]
        plt.imshow(np.asarray(image))
        plt.show()


    def get_label(self, index):
        '''
        Returns the indexed label from the training dataset
        '''
        return self.labels[index]

    def get_image(self, index):
        '''
        Returns the indexed label from the training dataset
        '''
        return self.data[index]

    def plot_distribution(self, dataset_type='training'):
        """
        Plot the distribution of labels in the training or validation dataset.
        Inputs:
            dataset_type: 'training' or 'validation'
        """
        unique_labels, counts = np.unique(self.labels, return_counts=True)

        plt.bar(unique_labels, counts, align='center', alpha=0.7)
        plt.xticks(unique_labels)
        plt.xlabel('Label')
        plt.ylabel('Total Count')
        plt.title(f'Distribution of Labels in {dataset_type} Dataset')
        plt.show()

```

2.6. Write your model architecture, you can be creative here! Here is a (non exhaustive) list of some useful documentations you could want to use [Conv2d](https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html), [Linear](https://pytorch.org/docs/stable/generated/torch.nn.Linear.html), [LayerNorm](https://pytorch.org/docs/stable/generated/torch.nn.LayerNorm.html), [activation functions](https://pytorch.org/docs/stable/nn.html#non-linear-activations-weighted-sum-nonlinearity). Justify your choice of architecture. **[5 pts]**

The following model Architecture consist of 3 convolutional layers, activation layers and pooling layers, with 2 fully connected layer, dropout layer and sigmoid activation.

2D CNN is chosen for convolutional layers as it is generally used in image analysis and is designed to extract features from images. As mentioned in Q1.5, CNN is memory and computationally-efficient. Each convolutional layer uses a 3x3 kernel to convolve the input.

For activation layers, although Leaky ReLU is more stable and does better job in mitigating dead neuron problems when compared to traditional ReLU, it is more computationally expensive. To save computation time, ReLU is instead chosen.

MaxPooling2D is used as pooling layers to max pooling with a 2x2 kernel and a stride of 2. It is designed to remove insignificant features and retain important features.

Fully connected layers allow extraction of complex and high-level features from convolutional and pooling layers, and produce final binary classification output. Therefore, two fully connected layers are utilized in the model. The first layer processes the 9216 features from the third convolutional and pooling layers, and reduces it to 128 neurons. The second layer processes the output from the first layer and produces the final 2 output features.

A dropout layer is added between the first fully connected layer and the second fully connected layer to prevent overfitting during training. The layer randomly deactivates a fraction of neurons during training, preventing the model from relying heavily on specific features, thereby enhancing generalization. The standard value of 0.5 is chosen as it is the convention and computationall efficient.

Finally, The sigmoid activation function, located after the second fully connected layer, is applied at the end of the model, which produces the final probablistic output values between 0 and 1. The model's output is interpreted as the likelihood of belonging to one of the two classes.


```python
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
#         # ****START CODE****

        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        self.fc1 = nn.Linear(64 * 12 * 12, 128)
        self.fc2 = nn.Linear(128, 2)
        self.dropout = nn.Dropout()
        self.sig = nn.Sigmoid()

    def forward(self, x):

        x = self.conv1(x)
        x = self.relu(x)
        x = self.pool(x)

        x = self.conv2(x)
        x = self.relu(x)
        x = self.pool(x)

        x = self.conv3(x)
        x = self.relu(x)
        x = self.pool(x)

        x = x.view(-1, 64 * 12 * 12)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        x = self.sig(x)

#         # ****END CODE****

        return x
```

```python
model = ConvNet().to(device)
```

```python
# ****START CODE****
from torchmetrics.classification import Accuracy
import torch.optim as optim

lr = 0.01
num_epochs = 50
BATCH_SIZE = 32
optimizer = optim.SGD(model.parameters(), lr=lr)
criterion = nn.BCEWithLogitsLoss()
metric = Accuracy(task="binary").to(device)

# ****END CODE****
```

Train model and validate it after each epoch. Feel free to use a GPU if you're training on colab to speed up your training.

```python
# ****START CODE****

train_accuracies = []
val_accuracies = []
train_losses = []
val_losses = []

print(f"  Learning Rate: {lr:.4f}, Batch Size: {BATCH_SIZE}, Number of layers: {3}")

for epoch in range(num_epochs):
    # Training
    model.train()
    train_loss = 0.0
    train_metric = 0.0

    for image, target in train_dataloader:
        image, target = image.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(image)
        loss = criterion(output[:,1], target.float())
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
        train_metric += metric(output[:,1], target.float())

    avg_train_loss = train_loss / len(train_dataloader)
    avg_train_metric = train_metric / len(train_dataloader)

    # Validation
    model.eval()
    val_loss = 0.0
    val_metric = 0.0

    with torch.no_grad():
        for image, target in val_dataloader:
            image, target = image.to(device), target.to(device)
            output = model(image)
            loss = criterion(output[:,1], target.float())

            val_loss += loss.item()
            val_metric += metric(output[:,1], target.float())

    avg_val_loss = val_loss / len(val_dataloader)
    avg_val_metric = val_metric / len(val_dataloader)

    train_accuracies.append(avg_train_metric)
    val_accuracies.append(avg_val_metric)
    train_losses.append(avg_train_loss)
    val_losses.append(avg_val_loss)

    # Print results
    print(f"Epoch {epoch + 1}/{num_epochs}:")
    print(f"  Train Loss: {avg_train_loss:.4f}, Train Accuracy: {avg_train_metric:.4f}")
    print(f"  Val Loss: {avg_val_loss:.4f}, Val Accuracy: {avg_val_metric:.4f}")

# ****END CODE****
```

### 2) Generator and Discriminator Models

3.2.1) We will now implement the backbone for the generator and the discriminator. You are asked to complete the code for the ResidualBlock of the generator backbone **[4 pt]**:

```python
##############################
#           RESNET
##############################


class ResidualBlock(nn.Module):
    def __init__(self, in_features):
        # ****START CODE****

        super(ResidualBlock, self).__init__()
        self.conv_block = nn.Sequential(
            nn.ReflectionPad2d(1),
            nn.Conv2d(in_features, in_features, kernel_size=3, stride=1),
            nn.InstanceNorm2d(in_features),
            nn.ReLU(inplace=True),
            nn.ReflectionPad2d(1),
            nn.Conv2d(in_features, in_features, kernel_size=3, stride=1),
            nn.InstanceNorm2d(in_features)
        )

        # ****END CODE****

    def forward(self, x):
        # ****START CODE****

        return x + self.conv_block(x)

        # ****END CODE****


class GeneratorResNet(nn.Module):
    def __init__(self, input_channel, n_blocks, filters, output_channel):
        super(GeneratorResNet, self).__init__()

        # Initial convolution block
        model = [
            nn.ReflectionPad2d(input_channel),
            nn.Conv2d(input_channel, filters, 7),
            nn.InstanceNorm2d(filters),
            nn.ReLU(inplace=True),
        ]
        in_features = filters

        # Downsampling
        for _ in range(2):
            filters *= 2
            model += [
                nn.Conv2d(in_features, filters, 3, stride=2, padding=1),
                nn.InstanceNorm2d(filters),
                nn.ReLU(inplace=True),
            ]
            in_features = filters

        # Residual blocks
        for _ in range(n_blocks):
            model += [ResidualBlock(filters)]

        # Upsampling
        for _ in range(2):
            filters //= 2
            model += [
                nn.Upsample(scale_factor=2),
                nn.Conv2d(in_features, filters, 3, stride=1, padding=1),
                nn.InstanceNorm2d(filters),
                nn.ReLU(inplace=True),
            ]
            in_features = filters

        # Output layer
        model += [nn.ReflectionPad2d(output_channel), nn.Conv2d(filters, output_channel, 7), nn.Tanh()]

        self.model = nn.Sequential(*model)

    def forward(self, x):
        return self.model(x)


##############################
#        Discriminator
##############################


class Discriminator(nn.Module):
    def __init__(self, input_channel, filters):
        super(Discriminator, self).__init__()

        def discriminator_block(in_filters, out_filters, normalize=True):
            """Returns downsampling layers of each discriminator block"""
            layers = [nn.Conv2d(in_filters, out_filters, 4, stride=2, padding=1)]
            if normalize:
                layers.append(nn.InstanceNorm2d(out_filters))
            layers.append(nn.LeakyReLU(0.2, inplace=True))
            return layers

        self.model = nn.Sequential(
            *discriminator_block(input_channel, filters, normalize=False),
            *discriminator_block(filters, filters * 2),
            *discriminator_block(filters * 2, filters * 4),
            *discriminator_block(filters * 4, filters *8),
            nn.ZeroPad2d((1, 0, 1, 0)),
            nn.Conv2d(filters *8, 1, 4, padding=1)
        )

    def forward(self, img):
        return self.model(img)
```

We will now implement the different loss functions used in CycleGANs:

```python
SOFT_FAKE_LABEL_RANGE =  [0.0, 0.3] # The label of fake label will be generated within this range.
SOFT_REAL_LABEL_RANGE = [0.7, 1.2] # The label of real label will be generated within this range.
```

The discriminator loss is defined by:
\begin{equation}
\mathcal{L}_{D} = \frac{1}{2} (\mathbb{E}_{y \sim p_{data}(y)}[(D_Y(y) - r_2)^2] + \mathbb{E}_{x \sim p_{data}(x)}[(D_Y(G_{XY}(x))-r_1)^2]) + \frac{1}{2} (\mathbb{E}_{x \sim p_{data}(x)}[(D_X(x) - r_2)^2] + \mathbb{E}_{y \sim p_{data}(y)}[(D_X(G_{YX}(y))-r_1)^2])
\end{equation}
with $p_{data}(x)$ being the distribution of images from the first domain, $p_{data}(y)$ being the distribution of images from the second domain, $G_{XY}$ and $G_{YX}$ being the two generators, $D_X$ and $D_Y$ the two discriminators, and $r_1$ and $r_2$ being the soft fake and real labels chosen from a uniform distribution within the ranges $[0.0, 0.3]$ and $[0.7, 1.2]$ respectively.

3.3.2. You will now implement the discriminator loss function **[4 pts]**:

```python
def discriminator_loss(real_image, generated_image):
    # ****START CODE****

    r1 = torch.rand_like(generated_image).uniform_(SOFT_FAKE_LABEL_RANGE[0], SOFT_FAKE_LABEL_RANGE[1])
    r2 = torch.rand_like(real_image).uniform_(SOFT_REAL_LABEL_RANGE[0], SOFT_REAL_LABEL_RANGE[1])

    loss_real = 0.5 * torch.mean((real_image - r2) ** 2)
    loss_fake = 0.5 * torch.mean((generated_image - r1) ** 2)

    loss = loss_real + loss_fake

    # ****END CODE****
    return loss
```

The generator loss is defined by:
\begin{equation}
\mathcal{L}_{G} = \mathbb{E}_{x \sim p_{data}(x)}(D_Y(G_{XY}(x)) - r_2)^2 + \mathbb{E}_{y \sim p_{data}(y)}(D_X(G_{YX}(y)) - r_2)^2
\end{equation}
with $p_{data}(x)$ being the distribution of images from the first domain, $p_{data}(y)$ being the distribution of images from the second domain, $G_{XY}$ and $G_{YX}$ being the two generators, $D_X$ and $D_Y$ the two discriminators, and $r_2$ being the soft real label chosen from a uniform distribution within the range $[0.7, 1.2]$.

3.2.3. You will now implement the generator loss function for a domain **[4 pts]**:

```python
def generator_loss(generated_image):
    # ****START CODE****

    r = torch.rand_like(generated_image).uniform_(SOFT_REAL_LABEL_RANGE[0], SOFT_REAL_LABEL_RANGE[1])

    loss = torch.mean((generated_image - r)**2)

    # ****END CODE****
    return loss
```

In addition to the traditional loss functions used in GANs, CycleGANs also use two additional loss functions: cycle consistency loss and identity loss. We will use the same $\lambda$ for the two losses.

```python
LAMBDA = 10
```

The cycle consistency loss is defined by:
\begin{equation}
\mathcal{L}_{cyc} = \lambda\mathbb{E}_{x \sim p_{data}(x)}[||x - G_{YX}(G_{XY}(x))||_1] + \lambda\mathbb{E}_{y \sim p_{data}(y)}[||y - G_{XY}(G_{YX}(y))||_1]
\end{equation}
with $p_{data}(x)$ being the distribution of images from the first domain, $p_{data}(y)$ being the distribution of images from the second domain, $G_{XY}$ and $G_{YX}$ being the two generators and $\lambda$ being the weight for the cycle consistency loss.

3.3.4. You will now implement the cycle consistency loss function **[4 pts]**:

```python
def cycle_consistency_loss(real_image, cycled_image):
    # ****START CODE****

    loss = LAMBDA * (torch.mean(torch.abs(real_image - cycled_image)))

    # ****END CODE****
    return loss
```

The identity loss is defined by:
\begin{equation}
\mathcal{L}_{id} = \frac{1}{2}\lambda\mathbb{E}_{x \sim p_{data}(x)}[||G_{YX}(x) - x||_1] + \frac{1}{2}\lambda\mathbb{E}_{y \sim p_{data}(y)}[||G_{XY}(y) - y||_1]
\end{equation}
with $p_{data}(x)$ being the distribution of images from the first domain, $p_{data}(y)$ being the distribution of images from the second domain, $G_{XY}$ and $G_{YX}$ being the two generators, and $\lambda$ being the weight for the identity loss.

3.3.5. You will now implement the identity loss function **[4 pts]**:

```python
def identity_loss(real_image, generated_image):
    # ****START CODE****

    loss = 0.5 * LAMBDA * torch.mean(torch.abs(generated_image - real_image))

    # ****END CODE****
    return loss
```

The total generator loss is defined by:
\begin{equation}
\mathcal{L}_{G_{tot}} = \mathcal{L}_{G} + \mathcal{L}_{cyc} + \mathcal{L}_{id}
\end{equation}
with $\mathcal{L}_{G_X}$ and $\mathcal{L}_{G_Y}$ being the generator loss for the two domains, $\mathcal{L}_{cyc}$ being the cycle consistency loss and $\mathcal{L}_{id}$ being the identity loss.

### 5) Theory Questions **[6 pts]** (2 pts each):
5.1. What are the key differences between a traditional GAN and a CycleGAN?

The key differences between a traditional GAN and CycleGAN mainly lies in their objectives and architectures. GANs are designed to generate unconditional, realistic images from random noise, relying on paired training data. CycleGAN, on the other hand, is designed to perform image-to-image translation with unpaired training data. Therefore, the different objectives affects the overall architectures of these two modellings.

Architectural-wise, GANs consist of a generator that creates images and a discriminator that determines the fakeness of the created images. Conversely, CycleGAN has two GANs, meaning it contains two generators that map images in both directions and two discriminators that determines the fakeness individually. The loss functions are also different as a result. GAN uses adversarial loss between a generator and a discriminator, while CycleGAN has two loss functions, adversarial loss and cycle-consistency loss. Cycle-consistency loss attempts to minimize the loss between the Image A and the generated image, and the Image B and the generated image, ensuring the image preservation during bidrectional transformations.

5.2. How does the architecture of a CycleGAN ensure the preservation of key features in the image translation process?

The integration of cycle-consistency loss into the architecture of a CycleGAN plays a vital role in preserving key features during the image translation process. This loss, which measures the bidirectional difference between the original image and the one reconstructed after passing through the generators, is a crucial component. By minimizing the cycle-consistency loss, the model encourages the generation of images that maintain consistency with their originals. As a result,  key features are effectively preserved throughout the image translation process.

5.3. Discuss the role of cycle consistency loss in CycleGAN. Why is it important?

Cycle-consistency loss is a pivotal component in CycleGAN, playing a major role in preserving key features during the image translation process. Its primary function is to encourage bidirectional consistency in translation, ensuring that the process is reversible, leading to meaningful and reversible image mappings. The loss also serves to minimize inconsistencies between the original and generated images, preserving the integrity of the content. Additionally, by penalizing deviations from the original image, it emphasizes the generation of images that closely resemble the original structures, thereby retaining important features.

Model

```python
# ****START CODE****

##############################
#           RESNET
##############################


class ResidualBlock(nn.Module):
    def __init__(self, in_features):
        # ****START CODE****

        super(ResidualBlock, self).__init__()
        self.conv_block = nn.Sequential(
            nn.ReflectionPad2d(1),
            nn.Conv2d(in_features, in_features, kernel_size=3, stride=1),
            nn.ReLU(inplace=True), ## Solution 2
            nn.InstanceNorm2d(in_features), ## Solution 2
            nn.ReflectionPad2d(1),
            nn.Conv2d(in_features, in_features, kernel_size=3, stride=1),
            nn.InstanceNorm2d(in_features)
        )

        # ****END CODE****

    def forward(self, x):
        # ****START CODE****

        return x + self.conv_block(x)

        # ****END CODE****


class GeneratorResNet(nn.Module):
    def __init__(self, input_channel, n_blocks, filters, output_channel):
        super(GeneratorResNet, self).__init__()

        # Initial convolution block
        model = [
            nn.ReflectionPad2d(input_channel),
            nn.Conv2d(input_channel, filters, 7),
            nn.InstanceNorm2d(filters),
            nn.ReLU(inplace=True),
        ]
        in_features = filters

        # Downsampling
        for _ in range(2):
            filters *= 2
            model += [
                nn.Conv2d(in_features, filters, 3, stride=2, padding=1),
                nn.InstanceNorm2d(filters),
                nn.ReLU(inplace=True),
            ]
            in_features = filters

        # Residual blocks
        for _ in range(n_blocks):
            model += [ResidualBlock(filters)]

        # Upsampling
        for _ in range(2):
            filters //= 2
            model += [
                nn.Upsample(scale_factor=2),
                nn.Conv2d(in_features, filters, 3, stride=1, padding=1),
                nn.InstanceNorm2d(filters),
                nn.ReLU(inplace=True),
            ]
            in_features = filters

        # Output layer
        model += [nn.ReflectionPad2d(output_channel), nn.Conv2d(filters, output_channel, 7), nn.Tanh()]

        self.model = nn.Sequential(*model)

    def forward(self, x):
        return self.model(x)


##############################
#        Discriminator
##############################


class Discriminator(nn.Module):
    def __init__(self, input_channel, filters):
        super(Discriminator, self).__init__()

        def discriminator_block(in_filters, out_filters, normalize=True):
            """Returns downsampling layers of each discriminator block"""
            layers = [nn.Conv2d(in_filters, out_filters, 4, stride=2, padding=1)]
            if normalize:
                layers.append(nn.InstanceNorm2d(out_filters))
            layers.append(nn.LeakyReLU(0.2, inplace=True))
            return layers

        self.model = nn.Sequential(
            *discriminator_block(input_channel, filters, normalize=False),
            *discriminator_block(filters, filters * 2),
            *discriminator_block(filters * 2, filters * 4),
            *discriminator_block(filters * 4, filters *8),
            nn.ZeroPad2d((1, 0, 1, 0)),
            nn.Conv2d(filters *8, 1, 4, padding=1)
        )

    def forward(self, img):
        return self.model(img)

# ****END CODE****
```
